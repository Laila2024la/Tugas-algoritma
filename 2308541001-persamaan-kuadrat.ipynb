```python
import numpy as np
 
# Fungsi untuk menghitung koefisien a, b dan c dari persamaan kuadrat y = ax^2 + bx + c
def hitung_koefisien(titik1, titik2, titik3):
    # Membuat matriks A dan vektor b berdasarkan tiga titik yang diberikan
    A = np.array([[titik1[0]**2,titik1[0], 1],[titik2[0]**2, titik2[0], 1], [titik3[0]**2, titik3[0],1]])
    b = np.array([titik1[1],titik2[1],titik3[1]])
    
    # Menghitung koefisien a, b, dan c
    koefisien = np.linalg.solve(A,b)
    return koefisien
# Masukkan tiga titik dari pengguna
print("Masukkan tiga titik untuk menentukan persamaan kuadrat (format: x y):")
titik1 = tuple(map(float,input("titik1: ").split()))
titik2 = tuple(map(float,input("titik2: ").split()))
titik3 = tuple(map(float,input("titik3: ").split()))

# Hitung koefisien persamaan kuadrat
a, b, c = hitung_koefisien(titik1, titik2, titik3)

# Cetak persamaan kuadrat
print(f"persamaan kuadrat yang melalui titik-titik tersebut adalah: y = {a}x^2 + {b}x + {c}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Koefisien persamaan kuadrat
a = 1
b = 2
c = 3

# Rentang nilai x
x = np.linspace(-10,10,400)

# Persamaan kuadrat
y = a * x**2 + b * x + c

# Membuat grafik
plt.figure(figsize=(8, 6))
plt.plot(x,y, label=f'y = {a}x^2 + {b}x + {c}')

# Menambahkan garis sumbu x dan y
plt.axhline(0,color='black',linewidth=0.5)
plt.axhline(0,color='black',linewidth=0.5)

# Menambahkan grid
plt.grid(color = 'gray',linestyle = '--',linewidth = 0.5)

# Menambahkan judul dan label sumbu
plt.title('Grafik persamaan garis kuadrat')
plt.xlabel('x')
plt.ylabel('y')

# Menampilkan legenda
plt.legend()

# Menampilkan grafik
plt.show()
```
